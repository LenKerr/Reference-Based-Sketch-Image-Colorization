```python
import cv2
import torch
import numpy as np
import torch.nn.functional as F
from torchvision import transforms as T
from PIL import Image
```

```python
def TPS(x):
    
    def affine_transform(x, theta):
        theta = theta.view(-1, 2, 3)
        grid = F.affine_grid(theta, x.size())
        x = F.grid_sample(x, grid)
        return x
    
    theta1 = np.zeros(9)
    theta1[0:6] = np.random.randn(6) * 0.15
    theta1 = theta1 + np.array([1,0,0,0,1,0,0,0,1])
    affine1 = np.reshape(theta1, (3,3))
    affine1 = np.reshape(affine1, -1)[0:6]
    affine1 = torch.from_numpy(affine1).type(torch.FloatTensor)
    x = affine_transform(x, affine1) # source image
    return x

to_tensor = T.ToTensor()
```

```python
import albumentations as A
from albumentations.pytorch import transforms as TA


def UniformNoise(x):
    return x + np.random.uniform(-50, 50, (256,256,3))

def TPS(x):
    
    def affine_transform(x, theta):
        theta = theta.view(-1, 2, 3)
        grid = F.affine_grid(theta, x.size())
        x = F.grid_sample(x, grid)
        return x
    
    theta1 = np.zeros(9)
    theta1[0:6] = np.random.randn(6) * 0.15
    theta1 = theta1 + np.array([1,0,0,0,1,0,0,0,1])
    affine1 = np.reshape(theta1, (3,3))
    affine1 = np.reshape(affine1, -1)[0:6]
    affine1 = torch.from_numpy(affine1).type(torch.FloatTensor)
    x = affine_transform(x, affine1) # source image
    return x

transform_common = []
transform_a = []
transform_s = []
mode='train'
crop_size = 256
image_size = 256

if mode == 'train':
    transform_common.append(A.HorizontalFlip(p=0.5))
    transform_common.append(A.ShiftScaleRotate())
#     transform_common.append(A.RandomResizedCrop(image_size, image_size))
    transform_common.append(A.Resize(crop_size, crop_size))
    transform_common.append(A.RandomCrop(image_size, image_size))
else:
    transform_common.append(A.Resize(image_size, image_size))

transform_a.append(T.ToPILImage())
transform_a.append(T.ColorJitter(brightness=0.2, contrast=0.3, saturation=0.2, hue=0.3))
transform_a.append(T.ToTensor())
transform_a.append(T.Normalize(mean=(0.5,0.5,0.5), std=(0.5,0.5,0.5)))

transform_common = A.Compose(
    transform_common)

transform_a = T.Compose(transform_a)

transform_a.append(T.ToPILImage())
transform_a.append(T.ToTensor())
transform_a.append(T.Normalize(mean=0.5, std=0.5))
```

```python
def ToPIL(x):
    min_, max_ = x.min(), x.max()
    x.add_(-min_).div_(max_ - min_ + 1e-5)

    return T.ToPILImage()(x)
```

```python
import os
from torch.utils import data
from glob import glob

class Dataset(data.Dataset):
    def __init__(self, transform_common, transform_a):
        
        self.image_dir = 'color/'
        self.line_dir = 'xdog/'
        
        self.transform_common = transform_common
        self.transform_a = transform_a
        self.transform_s = transform_s
        
        self.ids = [f.split('/')[-1] for f in glob(os.path.join(self.line_dir, '*'))]

    def __getitem__(self, index):
        # filename = self.ids[index] + '.png'
        filename = '17938.png'
        
        image_path = os.path.join(self.image_dir, filename)
        line_path = os.path.join(self.line_dir, filename)
        
        image = cv2.imread(image_path)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        line = cv2.imread(line_path, cv2.IMREAD_GRAYSCALE)
        
        aug = self.transform_common(image=image, mask=line)
        image = aug['image']
        line = aug['mask']
        print(image.shape, line.shape)
#         image += np.random.uniform(-50, 50, (256,256,3)).astype('uint8')

        image = self.transform_a(image)
        print(image.size())
        
        image_tps = TPS(image.unsqueeze(0))
        
        return image, image_tps, T.ToTensor()(line)
        
    def __len__(self):
        return len(self.ids)
```

```python
dataset = Dataset(transform_common, transform_a)
```
